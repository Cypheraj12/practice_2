#**REMOVING/FILLING NULL VALUES AND DUPLICATES**

```python
# Identify column types
numeric_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
categorical_cols = ['City', 'Date', 'AQI_Bucket']
```

```python
# --- Fill numeric columns with median ---
for col in numeric_cols:
    Raw_Data[col] = Raw_Data[col].fillna(Raw_Data[col].median())

# --- Fill categorical columns with mode ---
for col in categorical_cols:
    if Raw_Data[col].isnull().sum() > 0:
        Raw_Data[col] = Raw_Data[col].fillna(Raw_Data[col].mode()[0])
```

#**UNIVARIATE ANALYSIS**

```python
#@title 📈 DISTRIBUTION OF NUMERIC FEATURES (HISTOGRAMS)

plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    ax = plt.subplot((len(numeric_cols)//3)+1, 3, i)
    ax.set_facecolor('mintcream')
    sns.histplot(Raw_Data[col].dropna(), kde=True, bins=30, color='royalblue')
    plt.title(f'Distribution of {col}', fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

#REMOVING OUTLIER

```python
Q1 = Raw_Data["AQI"].quantile(0.25)
Q3 = Raw_Data["AQI"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers based on AQI
Raw_Data_no_outliers = Raw_Data[(Raw_Data["AQI"] >= lower_bound) & (Raw_Data["AQI"] <= upper_bound)]

print("Number of records before removing outliers:", len(Raw_Data))
print("Number of records after removing outliers:", len(Raw_Data_no_outliers))
```

#FEATURE ENGINEERING

```python
Raw_Data.reset_index(drop=True, inplace=True)



print("\n" + "="*70)
print("FEATURE ENGINEERING")
print("="*70)

# Extract selected basic time features
Raw_Data['Year'] = Raw_Data['Date'].dt.year

# Create season (India-specific)
Raw_Data['Season'] = (
    Raw_Data['Date'].dt.month.apply(lambda x: 'Winter' if x in [11,12,1,2,3]
                                    else 'Summer' if x in [4,5,6]
                                    else 'Monsoon' if x in [7,8,9]
                                    else 'Post-Monsoon')
)

# Categorize AQI
Raw_Data['AQI_Category'] = Raw_Data['AQI'].apply(
    lambda aqi: 'Unknown' if pd.isna(aqi)
    else 'Good' if aqi <= 50
    else 'Moderate' if aqi <= 100
    else 'Poor' if aqi <= 200
    else 'Very Poor' if aqi <= 300
    else 'Severe' if aqi <= 400
    else 'Hazardous'
)

# Pollution Severity Score (weighted average)
if all(col in Raw_Data.columns for col in ['PM2.5', 'PM10', 'NO2', 'SO2']):
    Raw_Data['Pollution_Score'] = (
        Raw_Data['PM2.5'] * 0.4 +
        Raw_Data['PM10'] * 0.3 +
        Raw_Data['NO2'].fillna(0) * 0.15 +
        Raw_Data['SO2'].fillna(0) * 0.15
    )

# Display created features
print("\n✅ New Features Created:")
new_features = ['Year','Season', 'AQI_Category', 'Pollution_Score']

for feature in new_features:
    if feature in Raw_Data.columns:
        print(f"   • {feature}")

print(f"\n📊 Updated Dataset Shape: {Raw_Data.shape}")
print("\n📋 Sample of new features:")
print(Raw_Data[['Date', 'Year', 'Season', 'AQI_Category','Pollution_Score']].head(10))
```
